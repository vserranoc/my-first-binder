```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=plt.imread('berlin-philharmonie.png')
```

```python
img
```

```python
img.shape
```

```python
img = img[:,:,0]
```

```python
plt.imshow(img, cmap='gray')
plt.title('Imagen en escala de grises')
plt.show()
```

```python
U,S,V_T = np.linalg.svd(img, full_matrices=False)
```

```python
print(U.shape)
print(S.shape)
print(V_T.shape)
```

```python
img_svd = (U*S)@V_T
```

```python
plt.imshow(img_svd, cmap='gray')
plt.title('Imagen utilizando SVD')
plt.show()
```

```python
(U[:,1]*S[1]).shape
```

```python
img_svd_1=np.outer((U[:,1]*S[1]),V_T[1,:])
```

```python
plt.imshow(img_svd_1, cmap='gray')
plt.title('SVD truncada a 1')
plt.show()
```

```python
img_svd_2=np.outer((U[:,1]*S[1]),V_T[1,:]) + np.outer((U[:,2]*S[2]),V_T[2,:])
```

```python
plt.imshow(img_svd_2, cmap='gray')
plt.title('SVD truncada a 2')
plt.show()
```

```python
k = 10
for i in range(k-1):
    temp = np.outer((U[:,i]*S[i]),V_T[i,:])
    img_svd_k += temp
```

```python
plt.imshow(img_svd_k, cmap='gray')
plt.title('SVD truncada a 10')
plt.show()
```
